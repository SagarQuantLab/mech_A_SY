**Agile**
- Projects are broken into smaller phases
- Strong focus on customer involvement and team work
- Embrace changes even late in development
- Regular reviews and adjustment to improve efficiency
- Prioritizes delievering usable features early and often
 
**Kanban**
- Work is delievered as soon as it's ready
- No roles defined (team member shared responsibility)
- Planning task are pulled when capacity allowed
- Change management - Changes can be intrdouced anytime
- Focus - Visualize workflow, limiting WIP and optimize flow

**Scrum**
- Fixed time sprint (2/4 week) with defined goal/with planned delieverable
- Defined roles - Product Owner, Scrum Master, Development Team
- Spring Planning before starting a sprint
- Changes are discourage mid-sprint, handled in next sprint/finish first then take changes
- Delievering stages of work

**Rules for naming variables**
- Variable names can only contain letters, digits and underscores (_).
- A variable name cannot start with a digit.
- Variable names are case-sensitive like `myVar` and `myvar` are different.
- Avoid using Python keywords like `if, else, for` as variable names.

**Rules followed by Software developer**
- Constant variable always upper case `PORT_NUMBER = 1000`
- Variable name should make sense Eg: `n_files` not `i`
- Classes name - PascalCase - `MyLifeMyRules`
- Function name - snake_case - `my_life_my_rules()`
- Protected function name - `_` at starting => `_my_life_my_rules()`
- Private function name - `__` at starting => `__my_life_my_rules()`
- Module & packages - use `-` => `scikit-learn` or `scikitlearn`
